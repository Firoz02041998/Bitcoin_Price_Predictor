```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("bitcoin_price_Training - bitcoin_price.2013Apr-2017Aug.csv.csv")
```

```python
test = pd.read_csv("bitcoin_price_1week_Test - bitcoin_price_Aug2017_1week.csv")
```

```python
test
```

```python
train
```

```python
train['Close'] = train['Close'].astype(float)
```

```python
train['Close'].plot(figsize=(40,20),title='Bitcoin Price')
```

```python
close_data = train.iloc[:,4:5].astype(float).values
```

```python
close_test = test.iloc[:,4:5].astype(float).values
```

```python
close_data
```

```python
close_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
close_data = scaler.fit_transform(close_data)
close_test = scaler.fit_transform(close_test)
```

```python
train_set = close_data
test_set = close_test
```

```python
x_train = train_set[0:len(train_set)-1]
y_train = train_set[1:len(train_set)]
```

```python
x_test = test_set[0:len(test_set)-1]
y_test = test_set[1:len(test_set)]
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
x_train = np.reshape(x_train,(len(x_train),1,x_train.shape[1]))
x_test = np.reshape(x_test,(len(x_test),1,x_test.shape[1]))
```

```python
model = Sequential()
```

```python
model.add(LSTM(256, return_sequences = True, input_shape = (x_train.shape[1],x_train.shape[2])))
model.add(LSTM(256))
model.add(Dense(1))
```

```python
model.compile(loss = 'mean_squared_error', optimizer='adam')
model.fit(x_train,y_train,epochs=50,batch_size=18,shuffle=False)
```

```python
predicted_price = model.predict(x_test)
predicted_price = scaler.inverse_transform(predicted_price)
real_price = scaler.inverse_transform(y_test)
```

```python
predicted_price
```

```python
real_price
```

```python
plt.figure(figsize=(40,5))
plt.plot(predicted_price, color='red', label= 'Predicted Price of Bitcoin ')
plt.plot(real_price, color='blue', label= 'Real Price of Bitcoin ')
plt.title('Predicted Price VS Real Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```

```python
model.evaluate(x_test,y_test)
```
